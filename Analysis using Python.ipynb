```python
df["year"] = df["order_date"].dt.year

df["month"] = df["order_date"].dt.month_name()
```

```python
df.head()
```

# Total sales per customer

```python
customer_sales = df.groupby("customer_name")["sales"].sum()
customer_sales.sort_values(ascending=False)
```

# Moving Average sales per month

```python
monthly_sales = df.groupby("month")["sales"].sum()
monthly_sales
```

```python
monthly_sales = df.groupby(df["order_date"].dt.to_period("M"))["sales"].sum()
monthly_sales
```

```python
moving_avg = monthly_sales.rolling(window=3).mean()
moving_avg
```

# Customers segmented based on total spending

```python
customer_total = df.groupby("customer_name")["sales"].sum()
customer_total
```

```python
def segment(x):
    
    if x < 5000:
        return "Low"
    
    elif x < 10000:
        return "Medium"
    
    else:
        return "High"


segments = customer_total.apply(segment)

print(segments)
```

# Top 10 Customers by Lifetime value

```python
top_10 = customer_total.sort_values(ascending=False).head(10)
top_10
```
